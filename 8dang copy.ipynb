```python
import pandas as pd
```

```python
r_data = pd.DataFrame()
for i  in range(12,23):
    tmp = pd.read_csv("C:/WORK/temp데이콘/data/rf_data/rf_20"+ str(i) +".csv")
    r_data=pd.concat([r_data, tmp])
r_data.reset_index().drop(labels="index", axis=1, inplace = True)
r_data
```

```python
w_data = pd.DataFrame()
for i  in range(12,23):
    tmp = pd.read_csv("C:/WORK/temp데이콘/data/water_data/data_20"+ str(i) +".csv")
    w_data=pd.concat([w_data, tmp])
w_data.reset_index().drop(labels="index", axis=1, inplace = True)
w_data
```

```python
r_data2 = r_data.drop(labels='ymdhm', axis=1)
r_data2
```

```python
data = pd.concat([w_data, r_data2], axis=1)
data
```

```python
train = data.iloc[:-24 * 6 * 48].drop(labels='ymdhm', axis = 1).reset_index(drop=True)
test = data.iloc[-24 * 6 * 48:].reset_index(drop=True)
```

```python

X = train[['swl', 'inf', 'sfw', 'ecpc', 'tototf', 'tide_level','fw_1018662', 'fw_1018680', 'fw_1018683','fw_1019630', 'rf_10184100', 'rf_10184110','rf_10184140']] 
y_1 = train[['wl_1018662']]
y_2 = train[['wl_1018680']]
y_3 = train[['wl_1018683']]
y_4 = train[['wl_1019630']]
```

```python
test_X = test[['swl', 'inf', 'sfw', 'ecpc', 'tototf', 'tide_level','fw_1018662', 'fw_1018680', 'fw_1018683','fw_1019630', 'rf_10184100', 'rf_10184110','rf_10184140']] 
```

```python
X.fillna(X.mean(),inplace=True)
y_1.fillna(y_1.mean(),inplace=True)
y_2.fillna(y_2.mean(),inplace=True)
y_3.fillna(y_3.mean(),inplace=True)
y_4.fillna(y_4.mean(),inplace=True)
```
